```python
import numpy as np
import pandas as pd
import tensorflow_datasets as tfds
import tensorflow as tf
```

```python
mnistdata,mnistinfo=tfds.load('mnist',as_supervised=True,with_info=True)
```

```python
countval=0.1*mnistinfo.splits['train'].num_examples
```

```python
countval=tf.cast(countval,tf.int64)
```

```python
countval
```

```python
counttest=mnistinfo.splits['test'].num_examples
```

```python
counttest=tf.cast(counttest,tf.int64)
```

```python
bufer=1000
```

```python
mnistdatatrain=mnistdata['train']
```

```python
mnistdatatest=mnistdata['test']
```

```python
mnistdatatrain=mnistdatatrain.shuffle(bufer)
```

```python
mnistdataval=mnistdatatrain.take(countval)
mnistdatabenchtrain=mnistdatatrain.skip(countval)
```

```python
for i,j in mnistdatabenchtrain.take(48):
    print(i,'---------',j)
```

```python
def scale(image,label):
    image=tf.cast(image,tf.float64)
    image/=255.
    return image,label
```

```python
mnistdataval=mnistdataval.map(scale)
mnistdatatest=mnistdatatest.map(scale)
mnistdatabenchtrain=mnistdatabenchtrain.map(scale)
```

```python
batch=100
```

```python
mnistdataval=mnistdataval.batch(countval)
```

```python
mnistdatatest=mnistdatatest.batch(counttest)
```

```python
mnistdatabenchtrain=mnistdatabenchtrain.batch(batch)
```

```python
valdata,valtarget=next(iter(mnistdataval))
```

```python
hidenlayer=50
```

```python
model=tf.keras.Sequential([
tf.keras.layers.Flatten(input_shape=(28,28,1)) ,
tf.keras.layers.Dense(hidenlayer,activation='relu') ,  
tf.keras.layers.Dense(hidenlayer,activation='relu'),
tf.keras.layers.Dense(hidenlayer,activation='relu'),
tf.keras.layers.Dense(hidenlayer,activation='relu'),
tf.keras.layers.Dense(10,activation='softmax')   
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(mnistdatabenchtrain,validation_data=(valdata,valtarget),epochs=6,verbose=2)
```

```python
model.evaluate(mnistdatatest)
```
