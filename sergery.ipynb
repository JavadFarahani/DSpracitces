```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as st
from sklearn.preprocessing import normalize as sn
import seaborn as sns
sns.set()
```

```python
n= 1000
```

```python
age=np.random.uniform(20,60,(n,1)).round(0)
```

```python
height=np.random.uniform(150,200,(n,1)).round(0)
```

```python
weight=np.random.uniform(50,130,(n,1)).round(0)
```

```python
smoke=np.random.random_integers(0,1,(n,1))
```

```python
health=np.random.random_integers(0,1,(n,1))
```

```python
x=np.column_stack((age,height,weight,smoke))
```

```python
x
```

```python
xtrain,xtest,ytrean,ytest=st(x,health,test_size=0.2,random_state=143)
```

```python
traindata=np.savez('traindata',input=xtrain,target=ytrean)
```

```python
testdata=np.savez('testdata',input=xtest,target=ytest)
```

```python
trdata=np.load('traindata.npz')
```

```python
tedata=np.load('testdata.npz')
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(1)])
compile=model.compile(loss='mean_squared_error')
model.fit(trdata['input'],trdata['target'],verbose=1,epochs=500)
```

```python
model.predict_on_batch(tedata['input']).round(5)
```
