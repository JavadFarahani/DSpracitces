```python
data=rawdata.copy()
```

```python
data['SAT']
```

```python
y=data['GPA']
x1=data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('GPA',fontsize=20)
plt.ylabel('Sat',fontsize=20)
plt.show()
sns.set()
```

```python
x=sts.add_constant(x1)
model=sts.OLS(y,x).fit()
model.summary()
```
