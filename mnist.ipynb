```python
import tensorflow_datasets as dsd
```

```python
mnistdata,mnistinfo=dsd.load(name='mnist',with_info=True,as_supervised=True)
```

```python
mnistinfo
```

```python
mtrain,mtest=mnistdata['train'],mnistdata['test']
```

```python
numvalidation=0.1*mnistinfo.splits['train'].num_examples
```

```python
numvalidation
```

```python
import tensorflow as tf
```

```python
numvalidation=tf.cast(numvalidation,dtype=tf.int64)
```

```python
numtest=mnistinfo.splits['test'].num_examples
```

```python
numtest=tf.cast(numtest,dtype=tf.int64)
```

```python
next(iter(mtrain))
```

```python
def scale(image,label):
    image=tf.cast(image,tf.float32)
    image /=255.
    return image,label    
```

```python
scaletrainandvalidation=mtrain.map(scale)
```

```python
scaletrainandvalidation
```

```python
scaletest=mtest.map(scale)
```

```python
bufersize=100
```

```python
shufledatatrainandvalidation=scaletrainandvalidation.shuffle(bufersize)
```

```python
traindata=shufledatatrainandvalidation.skip(numvalidation)
valdata=shufledatatrainandvalidation.take(numvalidation)
```

```python
batchsize=50
```

```python
traindata=traindata.batch(batchsize)
```

```python
valdata=valdata.batch(numvalidation)
```

```python
testdata=scaletest.batch(numtest)
```

```python
valinput,valtarget=next(iter(valdata))
```

```python
valdata
```

```python
valtarget
```

```python
valinput
```

```python
input=784
output=10
hidenlayersize=200
model=tf.keras.Sequential([
    
 tf.keras.layers.Flatten(input_shape=(28,28,1))   ,
 tf.keras.layers.Dense(hidenlayersize,activation='relu'),
 tf.keras.layers.Dense(hidenlayersize,activation='relu'),
 tf.keras.layers.Dense(hidenlayersize,activation='relu'),
 tf.keras.layers.Dense(output,activation='softmax')    
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
numepoch=10
```

```python
model.fit(traindata,epochs=numepoch,validation_data=(valinput,valtarget),verbose=1)
```

```python
model.evaluate(testdata)
```
