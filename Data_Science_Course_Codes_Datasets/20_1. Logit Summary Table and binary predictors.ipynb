```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('Binary Predictors.csv')
raw_data
```

```python
data = raw_data.copy()

data['Admitted'] = data['Admitted'].map({'Yes': 1, 'No': 0})
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})

data
```

## Logit Regression with different independent variables

```python
y = data['Admitted']
x0 = np.ones(168)

reg_log0 = sm.Logit(y,x0)

results_log0 = reg_log0.fit()

results_log0.summary()
```

```python
y = data['Admitted']
x1 = data['Gender']

x = sm.add_constant(x1)

reg_log = sm.Logit(y,x)

results_log = reg_log.fit()

results_log.summary()
```

```python
y = data['Admitted']
x1 = data[['SAT','Gender']]

x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```

```python
y = data['Admitted']
x1 = data[['SAT','Gender', 'Random']]

x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```
