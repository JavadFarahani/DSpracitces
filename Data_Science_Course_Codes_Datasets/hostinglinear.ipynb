```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression as lr
from sklearn.feature_selection import f_regression as fr
from sklearn.model_selection import train_test_split as ts
from sklearn.preprocessing import StandardScaler as ss
import seaborn as sns
```

```python
rawdata=pd.read_csv('housing.csv')
```

```python
data=rawdata.copy()
```

```python
data.head()
```

```python
data.describe()
```

```python
data.notna().sum()
```

```python
y=data['price']
```

```python
x=data.iloc[:,2:]
```

```python
x
```

```python
xstandard=ss()
```

```python
xstandard.fit(x)
```

```python
newx=xstandard.transform(x)
```

```python
xstandard
```

```python
fr(newx,y)[1].round(3)
```

```python
trainx,testx,trainy,testy=ts(x,y,test_size=0.3,random_state=123)
```

```python
model=lr()
```

```python
model.fit(trainx,trainy)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
mp=model.predict(testx).round(3)
```

```python
pdata=pd.DataFrame(testy)
```

```python
pdata=pdata.reset_index(drop=True)
```

```python
pdata
```

```python
pdata['predict']=mp
```

```python
pdata['predict']=pdata['predict'].astype(int)
```

```python
pdata
```

```python
pdata['Residual'] = pdata['price'] - pdata['predict']

pdata['Percent Error'] = np.absolute(pdata['Residual']/pdata['price']*100)

pdata
```
