```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_datasets as tfds
import ipywidgets as iw
```

```python
mnist,mnistinfo=tfds.load(name='mnist',with_info=True,as_supervised=True)
```

```python
mnistinfo
```

```python
countvaldition=mnistinfo.splits['train'].num_examples
```

```python
numtest=mnistinfo.splits['test'].num_examples
```

```python
countvaldition=0.1*countvaldition
```

```python
countvaldition=tf.cast(countvaldition,dtype=tf.int64)
```

```python
countvaldition
```

```python
mnisttrainval=mnist['train']
```

```python
mnisttest=mnist['test']
```

```python
mnisttrainval.take(5)
```

```python
mnisttrainval=mnisttrainval.shuffle(100)
```

```python
mylist=[]
for i in mnisttrainval.take(2):
   mylist.append(i) 
mylistframe=pd.DataFrame(mylist)
print(mylistframe)
```

```python
next(iter(mnisttrainval.take(5)))
```

```python
mnistval=mnisttrainval.take(countvaldition)
mnisttrain=mnisttrainval.skip(countvaldition)
```

```python
def scale (image,label):
    image=tf.cast(image,tf.float32)
    image/=255.
    return image,label
```

```python
mnisttrainval=mnisttrainval.map(scale)
```

```python
mnistval.map(scale)
mnisttrain.map(scale)
mnisttest.map(scale)
```

```python
bufersize=100
```

```python
mnisttrainval=mnisttrainval.shuffle(bufersize)
```

```python
valt,tart=next(iter(mnistval))
```

```python
input=784
output=10
hidenlayersize=200
model=tf.keras.Sequential([
    
 tf.keras.layers.Flatten(input_shape=(28,28,1))   ,
 tf.keras.layers.Dense(hidenlayersize,activation='relu'),
 tf.keras.layers.Dense(hidenlayersize,activation='relu'),
 tf.keras.layers.Dense(hidenlayersize,activation='relu'),
 tf.keras.layers.Dense(output,activation='softmax')    
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
batchsize=50
```

```python
mnisttrain=mnisttrain.batch(batchsize)
```

```python
mnistval=mnistval.batch(countvaldition)
```

```python
mnisttest=mnisttest.batch(numtest)
```

```python
numepoch=5
```

```python
mnistval
```

```python
tart
```

```python
valt
```

```python
model.fit(mnisttrain,validation_data=(valt,tart),epochs=numepoch,verbose=1)
```

```python
model.evaluate(mnisttest)
```
