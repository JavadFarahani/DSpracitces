## Logistic Regression by scikit-learn


```python
import pandas as pd

import numpy as np

from sklearn import metrics 

from sklearn.linear_model import LogisticRegression
```

```python
raw_data = pd.read_csv('Binary Predictors.csv')

raw_data
```

```python
df = raw_data.copy()

df['Admitted'] = df['Admitted'].map({'No': 0, 'Yes': 1})

df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

df = df.drop("Random", axis=1)

df
```

```python
x = df.drop('Admitted', axis = 1)

y = df.Admitted
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(x, y)
```

```python
x_test_raw = pd.read_csv('Test dataset.csv')

x_test = x_test_raw.drop('Admitted', axis = 1)

x_test['Gender'] = x_test['Gender'].map({'Male': 0, 'Female': 1})

y_test = x_test_raw['Admitted'].map({'No': 0, 'Yes':1})

x_test
```

```python
y_pred = logistic_regression.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)

accuracy_percentage = 100 * accuracy

accuracy_percentage, round(accuracy, 4)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_pred, y_test)
```
